### load libraries

```python
import json, os, subprocess
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
from PIL import Image
from PIL import ImageFont
from PIL import ImageDraw 
```

### write image with classified label and score

```python
for ss in range(1, 6):
    vidname = 'friend%d' % ss
    
    S = torch.load('%s.pth' % vidname)
    N = F.softmax(S, dim=1)
    V, I = N.max(dim=1)

    with open('output_%s.json' % vidname, 'r') as f:
        J = json.load(f)
        J = J[0]
        K = J['clips']

    folder_dir = '%s/' % vidname
    result_dir = '%s_result/' % vidname

    if not os.path.exists(result_dir):
        #subprocess.call('mkdir ' + result_dir, shell=True)
        subprocess.call('cp -r %s %s' % (folder_dir, result_dir), shell=True)

    for i in range(0, len(K)):
    #for i in range(0, 100):
        #S = torch.Tensor(K[0]['scores'])
        if(V[i]>0.4):
            print(i, K[i]['label'], V[i], K[i]['label_score'], K[i]['segment'])

            for j in range(K[i]['segment'][0],K[i]['segment'][1]):
                img = Image.open(folder_dir + "image_%05d.jpg" % j)
                draw = ImageDraw.Draw(img)
                font = ImageFont.truetype("calibri.ttf", 40)
                draw.text((0, 0),"[%0.2f] %s" % (V[i], K[i]['label']),(255,255,255),font=font)
                img.save(result_dir + "image_%05d.jpg" % j)
```

### FFMPEG encodes a sequence of images into a single video file

```python
for ss in range(1, 6):
    vidname = 'friend%d' % ss
    subprocess.call('~/local/ffmpeg/ffmpeg -i %s_result/image_%%05d.jpg -c:v libx264 -vf "fps=24,format=yuv420p" %s_out.mp4' % (vidname, vidname))
```

### write result json following the given format

```python
threshold = 0.5
for ss in range(1, 6):
    vidname = 'friend%d' % ss
    
    S = torch.load('%s.pth' % vidname)
    N = F.softmax(S, dim=1)
    V, I = N.max(dim=1)

    with open('output_%s.json' % vidname, 'r') as f:
        J = json.load(f)
        J = J[0]
        K = J['clips']

    folder_dir = '%s/' % vidname

    Q = []
    for i in range(0, len(K)):
        prev_val = (-2, '')
        
        if(V[i]>threshold):
            print(i, K[i]['label'], V[i], K[i]['label_score'], K[i]['segment'])
            
            # extend temporal value in Q
            if(len(Q) > 0):
                if(Q[-1][0]+1 == i and K[i]['label'] == Q[-1][1]):
                    C = Q[-1][4]

                    # update value in last Q
                    Q[-1][0] = i
                    Q[-1][2] = (Q[-1][2] * (C) + V[i]) / (C+1)
                    Q[-1][3][1] = K[i]['segment'][1]
                    
                    Q[-1][4] = Q[-1][4] + 1

                else:
                    Q.append([i, K[i]['label'], V[i], K[i]['segment'], 1]) 
            else:
                Q.append([i, K[i]['label'], V[i], K[i]['segment'], 1])
                
    with open('data/friends/friend_s01_e%02d.json' % ss, 'w') as f:
        L = []
        for q in Q:
            label = q[1]
            score = float(q[2].view(-1).numpy()[0])
            segment = q[3]
            #print(label, score, segment[0], segment[1])
            L.append({"class": label, "score": score, "time": [float(segment[0]/24.0), float(segment[1]/24.0)]})
        json.dump(L, f)                
```
